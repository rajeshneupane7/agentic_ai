```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Number of rows
n_rows = 5000

# -------------------------------
# Dataset 1: Milk production data
# -------------------------------
cow_ids = np.random.randint(0,20,n_rows)

milk_yield = np.random.uniform(0, 20, n_rows)            # 0-20 kg
fat = np.random.uniform(0, 5, n_rows)                   # 0-5%
protein = np.random.uniform(0, 5, n_rows)               # 0-5%
scc = np.random.randint(0, 200001, n_rows)              # 0-200000
conductivity = np.random.uniform(0, 10, n_rows)         # example range 0-10

milk_df = pd.DataFrame({
    "cow_id": cow_ids,
    "milk_yield": milk_yield,
    "fat": fat,
    "protein": protein,
    "scc": scc,
    "conductivity": conductivity
})

milk_df.to_csv("milk_data.csv", index=False)
print("milk_data.csv created.")

# -----------------------------------
# Dataset 2: Cow behavior data
# -----------------------------------
rumination_time = np.random.uniform(0, 1000, n_rows)    # 0-1000 min
water_intake = np.random.uniform(30, 200, n_rows)       # 30-200 liters
alerts = np.random.choice([0, 1], n_rows)               # 0 or 1 for alert

behavior_df = pd.DataFrame({
    "cow_id": cow_ids,
    "rumination_time": rumination_time,
    "water_intake": water_intake,
    "alerts": alerts
})

behavior_df.to_csv("behavior_data.csv", index=False)
print("behavior_data.csv created.")
```
